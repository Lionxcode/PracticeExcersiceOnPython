```
'''
Taller 9 - Mínimo palíndromo

Enunciado del problema
Dado un string S, reorganice las letras para crear el mínimo palíndromo lexicográfico. Un palíndromo es un string tal que se lee igual de derecha a izquierda y de izquierda a derecha.



Entrada
El string S


Salida
El mínimo palíndromo lexicográfico, o -1 si no se puede crear un palíndromo.


Ejemplos

Entrada Ejemplo 1

aabcc
Salida Ejemplo 1

acbca

Entrada Ejemplo 2

aabbbccc
Salida Ejemplo 2

-1
'''
s = input()
alf = {}
for car in s:
  if car in alf:
    alf[car] = alf[car] + 1
  else:
    alf[car] = 1

impar = []
for i in alf:
  if (alf[i] % 2) != 0:
    impar = impar + [i]
    alf[i] = alf[i] - 1

if (len(impar)) >= 2:
  print("-1")

else:
  abc ='abcdefghijklmnopqrstuvwxyz'
  S = str()
  for letra in abc:
    if letra in alf:
      cant = (alf[letra]) / 2
      S = S + (str(letra)*int(cant))
  
  if (len(impar) == 1): 
    mid = "".join(impar)
  else:
    mid = ""
  print(S + mid + S[::-1])

  
      
```

```
'''
Taller 9 - Espiral Personalizada

Enunciado del problema
Dada una matriz de tamaño NxN organizarla de tal forma que el primer elemento sea el más pequeño, el segundo elemento sea
el segundo más pequeño, etc., donde cada elemento se toma en sentido horario (de exterior a interior) formando un espiral.

Entrada
La primera línea será un número N, que representa la dimensión de la matriz a.


Las siguiente N líneas contendrán N números cada una, que son los elementos de la matriz.


Salida
Debe imprimir la matriz separada por espacios


Ejemplos

Entrada Ejemplo 1

3
2 5 12
22 45 55
1 6 8
Salida Ejemplo 1

1 2 5
45 55 6
22 12 8

Entrada Ejemplo 2

4
1 1 1 1
2 2 2 2
1 1 1 1
0 0 0 0
Salida Ejemplo 2

0 0 0 0
1 2 2 1
1 2 2 1
1 1 1 1

'''
N = int(input())
mdes = []
for n in range(N):
  mdes = mdes + input().split()
for n in range(len(mdes)):
  mdes[n] = int(mdes[n])
mdes.sort()

lim = len(mdes)
x = 0
y = -1
c1 = -1
c2 = N
c3 = 0
num = 0
step = 1

mord = [[]]*N
for i in range(len(mord)):
  mord[i] = ["Na"]*N

while num != lim:
  
  if (x == c1) and (y == c2):
    step = 2
  elif (x == c2) and (y == c2):
    step = 3
  elif (x == c2) and (y == c1):
    step = 4
  elif (x == c3) and (y == c1):
    c1 = c1 + 1
    c2 = c2 - 1
    c3 = c3 + 1
    step = 1
  
  if step == 1:
    y = y + 1
  elif step == 2:
    x = x + 1
  elif step == 3:
    y = y - 1
  elif step == 4:
    x = x - 1
    
  mord[x][y] = mdes[num]
  num = num +1  

for linea in mord:
  for colum in range(len(linea)):
    linea[colum] = str(linea[colum])
for linea in mord:
  print(" ".join(linea))
  
```

```
'''Taller 9 - Dulces con descuento

Enunciado del problema
Usted está en una tienda local y quiere comprar algunas barras de chocolate. Hay n barras en la tienda, la i-ésima de ellas cuesta ai monedas (y usted quiere comprarlas todas).


Tiene m cupones diferentes que le permiten comprar barras de chocolate. El cupón i-ésimo le permite comprar q_i barras de chocolate, y solo tiene que pagar por las qi−1 más caras (por lo tanto, la barra más barata de esas qi barras es gratuita).


Puede usar solo un cupón; Si usa el cupón i, debe elegir qi barras y comprarlas usando el cupón, y comprar todas las barras n−qi restantes sin ningún descuento.


Para decidir qué cupón elegir, desea saber cuál será la cantidad total mínima de dinero que debe pagar si utiliza cada uno de los cupones de manera óptima.



Entrada
La primera línea contiene un entero n (2≤n≤300000) - el número de barras de chocolate en la tienda.


La segunda línea contiene n enteros a1, a2, ..., an (1≤ai≤1000000000), donde ai es el costo de la i-ésima barra de chocolate.


La tercera línea contiene un número entero m (1≤m≤n − 1): el número de cupones que tiene.


La cuarta línea contiene m enteros q1, q2, ..., qm (2≤qi≤n), donde qi es el número de barras de chocolate que tiene que comprar con el i-ésimo cupón para que el más barato de ellos sea gratis.


Salida
Imprima m enteros separados por saltos de linea, donde el i-ésimo de ellos debe ser la cantidad mínima de dinero que debe pagar si compra qi barras con el i-ésimo cupón, y todas las barras restantes por su precio completo.


Ejemplos

Entrada Ejemplo 1

7
7 1 3 1 4 10 8
2
3 4
Salida Ejemplo 1

27
30
'''
nb = int(input())
b = input().split()
nc = int(input())
c = input().split()
for i in range(len(b)):
  b[i] = int(b[i])
b.sort()
for cupon in c:
  cup = int(cupon)
  v = sum(b) - b[-cup]
  print(v)
```

```
'''
Taller 9 - Suma 1000

Escriba un programa que lea secuencias de números enteros de la entrada para determinar si hay dos números en la
secuencia cuya suma es igual a 1000. Si la secuencia contiene los números, debe imprimir 'SI', si no lo hace, debe
imprimir 'NO'.



Entrada
La entrada será una lista de números enteros no negativos x_i tal que -20000000000 <= x_i <= 2000000000.".


Salida
Debe imprimir "SI" si hay dos números en la lista tal que su suma es 1000, o en caso contrario "NO".


Ejemplos

Entrada Ejemplo 1

123 456 789 1 10 20 500 999
Salida Ejemplo 1

SI

Entrada Ejemplo 2

100 200 300 400 500
Salida Ejemplo 2

NO
'''

n = input()
nl = n.split()
ev = False
for num in nl:
  f = str(1000-int(num))
  if f != "500":
    if f in nl:
      ev = True
      break
  else:
    if nl.count("500") > 1:
      ev = True
      break
    else:
      ev = False
if ev:
  print("SI")
else:
  print("NO")

print(nl)
```

```
'''
Taller 9 - Parejas de XOR = 0

Enunciado del problema
Dado una lista a con n números, calcular la cantidad de parejas a_i, a_j tal que a_i XOR a_j = 0 y j > i.



Entrada
La primera linea contiene un entero N que es la cantidad de números en la lista.


La segunda línea contiene una lista de N números separados por espacios.


Salida
La cantidad de parejas encontradas.


Ejemplos

Entrada Ejemplo 1

5
1 3 1 4 3
Salida Ejemplo 1

2

Entrada Ejemplo 2

7
1 1 2 2 2 1 1
Salida Ejemplo 2

9
'''
N = int(input())
Ns = input()
Nl = Ns.split()
Nc = set(Nl)
parejas = int()
for num in Nc:
  cant = Nl.count(num)
  parejas = parejas + ((cant*(cant-1))/2)
  
print(int(parejas))
```

```
def process_input(lst):
  lst.sort()
  for i in range(len(lst)-1):
    if lst[i][0] == lst[i+1][0]:
      lst[i][1] = int(lst[i][1]) + int(lst[i+1][1])
      lst[i+1] = ["eliminar"]
  result = []
  for obj in lst:
    if obj != ["eliminar"]:
      obj[1] = int(obj[1])
      result = result + [obj]
  return result

def duplicate_inventory(inv):
  inv_2 = inv
  for obj in range(len(inv_2)):
    inv_2[obj][1] = inv_2[obj][1] * 2
  return inv_2
    


s = input()
lines = []
while s != 'FIN':
    lines.append(s.split(" "))
    s = input()
inventory1 = process_input(lines)
inventory2 = duplicate_inventory(inventory1)

print((inventory1))
print((inventory2))
```
