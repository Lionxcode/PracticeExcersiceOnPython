```
'''Calcular PAPA
Enunciado del problema
Diego acaba de finalizar su primer semestre en la universidad y está interesado en saber cuál será su promedio aritmético ponderado acumulado (PAPA). Hacer un programa 
que reciba las notas y cŕeditos de cada una de las 5 materias que Diego
cursó este semestre y retorne el valor de su PAPA mostrando exactamente dos decimales. Ver la sección de notas si tiene dudas sobre cómo calcular este promedio.


Entrada
La entrada está conformada por 10 números, cada uno en una línea diferente: los primeros dos números representan la nota y los créditos de la primera materia, en ese
mismo orden. Las siguientes 4 parejas de números siguen el mismo formato 
para las siguientes 4 materias. Las notas pueden ser números decimales. Los créditos son todos números enteros.


Salida
Imprimir un número con exactamente 2 decimales, el valor del PAPA de Diego.


Ejemplos

Entrada Ejemplo 1

4.3
3
5.0
4
3.5
4
4.9
4
2.7
2
Salida Ejemplo 1

4.23

Entrada Ejemplo 2

3.6
5
4.1
5
1.8
5
1.6
1
2.0
2
Salida Ejemplo 2

2.95

Notas
El promedio acumulado se calcula multiplicando la nota de cada materia por su cantidad de créditos, sumando todos estos productos y dividiendo al final por la suma 
total de créditos. Más formalmente:

PAPA=∑Mi=1ni∗ci∑Mi=1ci

, donde:

ni= nota de la i-sima materia.
ci= créditos de la i-sima materia.
M = cantidad total de materias.'''

I=float(input())
Ic=int(input())
II=float(input())
IIc=int(input())
III=float(input())
IIIc=int(input())
IV=float(input())
IVc=int(input())
V=float(input())
Vc=int(input())

print(format(((I*Ic)+(II*IIc)+(III*IIIc)+(IV*IVc)+(V*Vc))/(Ic+IIc+IIIc+IVc+Vc),'.2f'))
```

```
'''Perímetro y área del círculo
Enunciado del problema
Dado el radio r de un círculo calcular su perímetro y área, luego imprimir los valores en un mensaje como el siguiente:


Perímetro del círculo con radio r: perimetro
Área del círculo con radio r: area

Reemplazar r, perimetro y area por sus respectivos valores.


Entrada
Un número decimal positivo, el valor del radio r.


Salida
Imprimir un mensaje como el mostrado en el enunciado. Los valores de perímetro y área deben imprimirse con exactamente dos números decimales. El radio debe imprimirse
tal y como se leyó.


Ejemplos

Entrada Ejemplo 1

1.0
Salida Ejemplo 1

Perímetro del círculo con radio 1.0: 6.28
Área del círculo con radio 1.0: 3.14

Entrada Ejemplo 2

86.52
Salida Ejemplo 2

Perímetro del círculo con radio 86.52: 543.62
Área del círculo con radio 86.52: 23517.05

Notas
Para usar el valor de pi en python se puede usar la libreria math. Primero poner la sentencia import math y luego usar la variable math.pi donde sea necesaria.'''

import math

r=float(input())
print("Perímetro del círculo con radio",str(r)+":", format(2*math.pi*r,'.2f'))
print("Área del círculo con radio",str(r)+":", format(math.pi*r**2,'.2f'))      
```

```
'''Ecuación rectángulo
Enunciado del problema
El largo de un rectángulo es (x + a) y su ancho (x + b). Dados los valores a, b y el área del rectángulo, calcular el valor de x.


Entrada
La entrada está conformada por tres números enteros, cada uno en una línea diferente: los valores a, b y el área del rectángulo, en ese mismo orden.


Salida
Imprimir un número entero, el valor de x, sin decimales (La solución a todos los casos es entera).


Ejemplos

Entrada Ejemplo 1

12
14
840
Salida Ejemplo 1

16

Entrada Ejemplo 2

15
18
550
Salida Ejemplo 2

7
'''

import math

a=int(input())
b=int(input())
c=int(input())

print(int((-b-a+math.sqrt(a**2-(2*(a*b))+b**2+4*c))/2))
```

```
'''Promedio de temperaturas
Enunciado del problema
Juliana está interesada en saber la temperatura promedio de su ciudad en estos días. Para hacerlo tomó tres medidas diferentes, sin embargo cada una está en una unidad
de temperatura diferente. La primera está en grados Celsius (℃), la segunda en grados Fahrenheit (℉) y la tercera en Kelvin (K). Escriba un programa que reciba las
tres temperaturas e imprima un mensaje como el siguiente:


Promedio en grados Celsius: promedio_c
Promedio en grados Fahrenheit: promedio_f
Promedio en Kelvin: promedio_k

Reemplazar las 3 variables de promedio por su respectivo valor.


Entrada
La entrada está conformada por 3 números decimales cada uno en una línea diferente. El primer valor es la temperatura del primer día en grados Celsius, el segundo es
la temperatura del segundo día en grados Fahrenheit y el tercero es la tempera del tercer día en Kelvin.


Salida
Imprimir un mensaje de la forma mostrada en el enunciado. Cada uno de los promedios debe ser impreso con exactamente 1 número decimal. Se recomienda usar la tabla de 
conversiones mostrada más abajo.


Ejemplos

Entrada Ejemplo 1

18.0
66.4
290.15
Salida Ejemplo 1

Promedio en grados Celsius: 18.0
Promedio en grados Fahrenheit: 64.5
Promedio en Kelvin: 291.2

Entrada Ejemplo 2

-17.5
43.3
275.9
Salida Ejemplo 2

Promedio en grados Celsius: -2.8
Promedio en grados Fahrenheit: 26.9
Promedio en Kelvin: 270.3'''

c=float(input())
f=float(input())
k=float(input())

print("Promedio en grados Celsius:", format((c+(5*(f-32)/9)+(k-273.15))/3,'.1f'))
print("Promedio en grados Fahrenheit:", format(((9*c/5+32)+f+(9*(k-273.15)/5+32))/3,'.1f'))
print("Promedio en Kelvin:", format(((c+273.15)+(5*(f-32)/9+273.15)+k)/3,'.1f'))
```
